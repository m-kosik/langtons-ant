```python
import math
import numpy as np
import re

code = '>987v>.v\nv456<  :\n>321 ^ _@'
code = '>25*"!dlroW olleH":v\n                v:,_@\n                >  ^'
code = '01->1# +# :# 0# g# ,# :# 5# 8# *# 4# +# -# _@'
code = '2>:3g" "-!v\  g30          <\n |!`"&":+1_:.:03p>03g+:"&"`|\n @               ^  p3\\" ":<\n2 2345678901234567890123456789012345678'
print(code)


# Find the longest line in code (determine width of grid)
new_lines = [m.end() for m in re.finditer('\n', code)]
new_lines = [0] + new_lines
new_lines = new_lines + [len(code)]
diffs = [x - new_lines[i - 1] for i, x in enumerate(new_lines)][1:]
length_of_line = np.max(diffs)


missing_zeros = np.array([length_of_line]*len(diffs)) - np.array(diffs)
print(missing_zeros)
new_lines[0] = 1
code1 = ''.join( code[new_lines[x]-1:new_lines[x+1]-1] + missing_zeros[x]*'N' for x in range(len(diffs)-1) )
code = code1 + code[new_lines[len(diffs)-1]-1:] + (missing_zeros[len(diffs)-1])*'N'
print(code)

# Fill up the code string to max width in each line by padding with spaces, 
# i.e. make a full rectangle (for correct up/down iterations):
        
stack = []
direction = "right"
x = 0
output = ''

while x < len(code):
    print(stack)
    print(output)
    print('.......')
    if code[x] == "0": stack.append(0)
    elif code[x] == "1": stack.append(1)
    elif code[x] == "2": stack.append(2)
    elif code[x] == "3": stack.append(3)
    elif code[x] == "4": stack.append(4)
    elif code[x] == "5": stack.append(5)
    elif code[x] == "6": stack.append(6)
    elif code[x] == "7": stack.append(7)
    elif code[x] == "8": stack.append(8)
    elif code[x] == "9": stack.append(9)

    elif code[x] == "+":
        a = stack.pop()
        b = stack.pop()
        stack.append(a+b)

    elif code[x] == "-":
        a = stack.pop()
        b = stack.pop()
        stack.append(b-a)

    elif code[x] == "*":
        a = stack.pop()
        b = stack.pop()
        stack.append(a*b)

    elif code[x] == "/":
        a = stack.pop()
        if not a:
            stack.append(0)
        else:
            b = stack.pop()
            stack.append(math.floor(b/a))
            
    elif code[x] == "\%":
        a = stack.pop()
        b = stack.pop()
        stack.append(a%b)
    
    elif code[x] == "!":
        a = stack.pop()
        if a: stack.append(0)
        else: stack.append(1)
    
    elif code[x] == "`":
        a = stack.pop()
        b = stack.pop()
        stack.append(int(bool(b>a)))

    elif code[x] == "\"":
        if direction == "right":
            y = x + 1
            while code[y]!= "\"":
                stack.append(ord(code[y]))
                y = y + 1
        elif direction == "left":
            y = x - 1
            while code[y]!= "\"":
                stack.append(ord(code[y]))
                y = y - 1
        elif direction == "down":
            y = x + length_of_line
            while code[y]!= "\"":
                stack.append(ord(code[y]))
                y = y + length_of_line
        elif direction == "up":
            y = x - length_of_line
            while code[y]!= "\"":
                stack.append(ord(code[y]))
                y = y - length_of_line
        x = y
    
    elif code[x] == ">":
        direction = "right"
        
    elif code[x] == "<":
        direction = "left"
        
    elif code[x] == "v":
        direction = "down"
        
    elif code[x] == "^":
        direction = "up"        
        
    elif code[x] == "?":
        choice = np.random.randint(4)
        if choice == 0:
            direction = "right"
        if choice == 1:
            direction = "left"
        if choice == 2:
            direction = "down"
        if choice == 3:
            direction = "up"
               
    elif code[x] == "_":
        if not bool(len(stack)):
            direction = "right"
        else:
            a = stack.pop()
            if a:
                direction = "left"
            else:
                direction = "right"
    
    elif code[x] == "|":
        a = stack.pop()
        if a:
            direction = "up"
        else:
            direction = "down"
            
    elif code[x] == ":":
        if bool(len(stack)):
            a = stack.pop()
            stack.append(a)
            stack.append(a)
        else:
            stack.append(0)
        
    elif code[x] == "\\":
        a = stack.pop()
        b = stack.pop()
        stack.append(a)
        stack.append(b)
    
    elif code[x] == "$":
        a = stack.pop()
        
    elif code[x] == ".":
        a = stack.pop()
        output = output + str(a)
    
    elif code[x] == ",":
        print('and here')
        a = stack.pop()
        output = output + chr(a)
    
    elif code[x] == "#":
        print('hashtag')
        if direction == "right":
            x = x + 1
        elif direction == "left":
            x = x - 1
        elif direction == "down":
            x = x + length_of_line
        elif direction == "up":
            x = x - length_of_line
            
    elif code[x] == "p":
        y = stack.pop()
        t = stack.pop()
        v = stack.pop()
        # put element v at row y and column x: [y,x] = v
        code = code[:y*length_of_line+t] + chr(v) + code[y*length_of_line+t+1:]
        print('after put')
        print(code)
            
    elif code[x] == "g":
        y = stack.pop()
        t = stack.pop()
        # get element v at row y and column x: [y,x] = v
        v = code[y*length_of_line + t]
        stack.append(ord(v))
        
    elif code[x] == "@":
        print(output)
        break
            
    if direction == "right":
        x = x + 1
    elif direction == "left":
        x = x - 1
    elif direction == "down":
        x = x + length_of_line
    elif direction == "up":
        x = x - length_of_line
        
        
    

            
                
print(stack)
#output = "".join([ str(i) for i in stack])
print(output)

```
